ARGBEAST Debugging
==================

Theory
------

We define a restricted ARG $A=(C,M,R)$ where $C$ is a phylogenetic tree representing the _clonal frame_, $M$ is a map identifying the number and locations of regions on the sequence alignment $S$ affected by the recombination/conversion events, and $R$ is a set of recombinant edges which depart from and attach to points on the clonal frame.

The parameters of the model are the recombination rate parameter $\rho$, the tract length parameter $\delta$ and the population size parameter $\theta$.

Our MCMC algorithm seeks to sample from the joint posterior density
\begin{equation}
P(A,\theta,\rho,\delta|S) \propto P(S|A)P(A|\theta,\rho,\delta)P(\theta,\rho,\delta)
\end{equation}

To test our sampler implementation, however, we sample instead from
\begin{equation}
P(A|\theta,\rho,\delta) = P(R|M,C,\theta)P(M|C,\rho,\delta)P(C|\rho,\delta)
\end{equation}

Full two-taxon comparison
-------------------------

Firstly, I've simulated $10^5$ two-taxon ARGs with $\rho=5$ and $\delta=50$ under the model.  I've also used the add/remove operator in conjunction with the ARG scaling operator to produce a chain of ARGs of length $10^8$ with the same parameters. The first 10% of the MCMC chain is discarded to ensure we are past burn-in.

```python
def removeBurnin(df, frac=0.1):
    n = shape(df)[0]
    return df.tail(int(n*(1.0-frac)))

dfs = read_table('simulated.stats')
df = removeBurnin(read_table('addRemoveScaleTest.stats'))
```

The following shows only the first $10^5$ steps following the 10%, demonstrating that we have adequate convergence.

```python
plot(df.head(1000)['Sample'], df.head(1000)['arg.CFheight'])
```

The CF height displays very little correlation between values separated by relatively small numbers of steps, meaning that we should have a very high ESS from our chain:

```python
from pandas.tools.plotting import autocorrelation_plot
autocorrelation_plot(df.head(500)['arg.CFheight'])
```

The clonal frame height distribution should simply be its distribution under the coalescent with a constant population size of 1. That is

\begin{equation}
    P(t_{CF})=e^{-t_{CF}}
\end{equation}

```python
bins=arange(0,10,0.1)
dfs['arg.CFheight'].hist(bins=bins,normed=True, alpha=0.5, label="Simulated")
df['arg.CFheight'].hist(bins=bins, normed=True, alpha=0.5, label="MCMC")
plot(bins, exp(-bins), label=r"Truth ($e^{-t_{CF}}$)")
xlim((-.1,8))
xlabel('CF Height')
ylabel('Density')
legend()
```

The distribution of recombination counts also shows a difference.  Note that the true distribution here is similar to the CF height distribution, as the number of recombinations is drawn from a Poisson distribution with mean $2\rho t_{CF}$.

```python
bins = arange(0,40,1)
dfs['arg.nRecomb'].hist(normed=True, bins=bins, alpha=0.5, label="Simulation")
df['arg.nRecomb'].hist(normed=True, bins=bins, alpha=0.5, label="MCMC")
legend()
title("Recombination count")
```

```python
bins = arange(0,150,1)
dfs['arg.meanTractLength'].hist(normed=True, bins=bins, alpha=0.5, label="Simulation")
df['arg.meanTractLength'].hist(normed=True, bins=bins, alpha=0.5, label="MCMC")
legend()
title("Tract length")
```

```python
bins = arange(0,10000,100)
dfs['arg.meanInterTractLength'].hist(normed=True, bins=bins, alpha=0.5, label="Simulation")
df['arg.meanInterTractLength'].hist(normed=True, bins=bins, alpha=0.5, label="MCMC")
legend()
title("Inter-tract length")
```

The distributions of mean edge length and departure height also differ, although I don't have analytical expressions for the truth there.

```python
bins=arange(0,4,.05)
dfs['arg.meanDepartureHeight'].hist(normed=True,
                                    bins=bins,
                                    alpha=0.5, label="Simulation")
df['arg.meanDepartureHeight'].hist(normed=True,
                                   bins=bins,
                                   alpha=0.5, label="MCMC")
legend()
title("Recombinant edge departure heights")
```

```python
bins=arange(0,3,.05)
dfs['arg.meanEdgeLength'].hist(normed=True,
                                    bins=bins,
                                    alpha=0.5, label="Simulation")
df['arg.meanEdgeLength'].hist(normed=True,
                                   bins=bins,
                                   alpha=0.5, label="MCMC")
legend()
title("Recombinant edge lengths")
```

Two taxon comparison with fixed clonal frame
--------------------------------------------

With the clonal frame fixed, we sample from
\begin{equation}
P(R,M|C,\delta,\rho,\theta) = P(R|M,\theta)P(M|C,\rho,\delta)
\end{equation}

I've again run $10^5$ simulations and $10^8$ MCMC steps.  The tree used is

```python
from Bio import Phylo
from StringIO import StringIO
tree = Phylo.read(StringIO("(1:0.799841931948889,2:0.799841931948889):0.0;"), "newick")
Phylo.draw(tree)
```

```python
dfs = read_table('simulated_fixedCF.stats')
df = removeBurnin(read_table('addRemoveTest_fixedCF.stats'))
```

Now everything lines up nicely.
